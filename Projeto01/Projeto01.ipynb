```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('Moon.jpg', cv2.IMREAD_GRAYSCALE)
# plt.imshow(image)
# plt.show()
image.shape
```

```python
def amplifier(img, k):
    d1 = img.shape[0] * k; d2 = img.shape[1] * k
    enlarged_matrix = np.zeros((d1,d2), dtype=img.dtype)
    for r in range(img.shape[0]):
        for c in range(img.shape[1]):
            pixel = img[r,c]
            r_enlarged = r * k
            c_enlarged = c * k
            for i in range(k):
                for j in range(k):
                    enlarged_matrix[r_enlarged + i][c_enlarged + j] = pixel
    return enlarged_matrix        
```

```python
def reducer(img, k):
    d1 = img.shape[0] // k; d2 = img.shape[1] // k
    reduced_matrix = np.zeros((d1,d2), dtype=img.dtype)
    for r in range (reduced_matrix.shape[0]):
        for c in range(reduced_matrix.shape[1]):
            reduced_matrix[r,c] = img[r * k, c * k]
    return reduced_matrix
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,8))
axes = axes.flatten()
reduced_img = reducer(image, 10)
axes[0].imshow(image, cmap='gray'); axes[0].set_title("Imagem Original")
axes[1].imshow(reduced_img, cmap='gray'); axes[1].set_title("Imagem Reduzida")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,8))
axes = axes.flatten()
amplified_img = amplifier(reduced_img, 10)

axes[0].imshow(reduced_img, cmap='gray'); axes[0].set_title("Imagem Reduzida")
axes[1].imshow(amplified_img, cmap='gray'); axes[1].set_title("Imagem Reampliada")
plt.tight_layout()
plt.show()
```
